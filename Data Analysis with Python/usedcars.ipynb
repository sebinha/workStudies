```python
anova_results_2=stats.f_oneway(cars_anova_grp.get_group("honda")["price"], cars_anova_grp.get_group("jaguar")["price"])
```

 # Significant if F is greater than 1 and pvalue is lower than 0.05, it means the price between then are significantly different

# Pearson Correlation
If near to +1 large positive correlation, if near to -1 large negative correlation and if near to 0 means no relation.
# p-value
lesser than 0.001 - Strong certainty in the result, lesser than 0.05 moderate certainty, lesser than 0.1 weak certainty and greater than 1 no certainty
